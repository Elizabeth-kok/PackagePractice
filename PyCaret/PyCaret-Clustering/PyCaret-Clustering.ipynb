```python
import pandas as pd
df=pd.read_csv("Mall_Customers.csv")
df.head()
```

### Train Model

```python
from pycaret.clustering import setup,create_model,evaluate_model,predict_model,save_model,load_model
```

```python
s=setup(df)
```

```python
model=create_model("kmeans")
```

### Evaluate Model

```python
evaluate_model(model)
```

### Predict Model

```python
predict_model(model,data=df)
```

```python
import matplotlib.pyplot as plt
plt.hist(predict_model(model,data=df)["Cluster"])
plt.show()
```

### Save model

k-Means Clustering is an unsupervised learning algorithm that is used for clustering whereas KNN is a supervised learning algorithm used for classification

```python
save_model(model,"kmeans")
```

### Load Model

```python
model=load_model("kmeans")
```
