```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("train.csv")
df.head()
```

```python
print(df.Survived.value_counts())
```

```python
from pycaret.classification import setup,compare_models,evaluate_model,predict_model,save_model,load_model
```

### Train model

```python
s=setup(df,target="Survived")
```

```python
best_model=compare_models()
```

```python
best_model
```

```python
evaluate_model(best_model)
```

### Test Model with test data

```python
test=pd.read_csv("test.csv")
test.head()
```

```python
predict_model(best_model,test)
```

### Save model

```python
save_model(best_model,"logistic")
```

### Load Model

```python
model=load_model("logistic")
```
