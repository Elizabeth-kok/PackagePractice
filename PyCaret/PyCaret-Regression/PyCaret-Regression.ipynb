```python
import pandas as pd
```

```python
df = pd.read_csv("day.csv")
df.head()
```

```python
df.describe()
```

### Train Model

```python
from pycaret.regression import setup,compare_models,evaluate_model,predict_model,save_model,load_model
```

```python
s=setup(df, target="cnt")
```

```python
best_model=compare_models()
```

```python
best_model
```

### Evaluate model

```python
evaluate_model(best_model)
```

### Predict Model

```python
testdf=pd.read_csv("day_test.csv")
testdf.head()
```

```python
predict_model(best_model,data=testdf)
```

### Save model

```python
save_model(best_model,"linear")
```

### Load model

```python
model=load_model("linear")
```
